```r
# 상대경로로 데이터 파일 읽기
# 현재 노트북은 main/ 폴더에 있으므로, data/ 폴더로 올라가서 파일을 읽음

# 파일 존재 확인
file_path <- "../data/global_air_quality_dataset.csv"
print(paste("파일 경로:", file_path))
print(paste("파일 존재:", file.exists(file_path)))

# 데이터 읽기
if(file.exists(file_path)) {
  air_quality <- read.csv(file_path)
  print("데이터 로드 성공!")
  print(paste("데이터 행 수:", nrow(air_quality)))
  print(paste("데이터 열 수:", ncol(air_quality)))
  print("처음 5행 미리보기:")
  head(air_quality)
} else {
  print("파일을 찾을 수 없습니다. 경로를 확인해주세요.")
}
```

```r
# 데이터 분석 시작

# 데이터 구조 확인
print("데이터 구조:")
str(air_quality)

print("\n데이터 요약:")
summary(air_quality)

# AQI 등급 분류 함수
classify_aqi <- function(aqi) {
  if (aqi <= 50) return("좋음")
  else if (aqi <= 100) return("보통")
  else if (aqi <= 150) return("민감군 영향")
  else if (aqi <= 200) return("나쁨")
  else if (aqi <= 300) return("매우 나쁨")
  else return("위험")
}

# AQI 등급 추가
air_quality$AQI_Category <- sapply(air_quality$AQI, classify_aqi)

# 도시별 평균 AQI
city_avg_aqi <- aggregate(AQI ~ City, data = air_quality, FUN = mean)
city_avg_aqi <- city_avg_aqi[order(city_avg_aqi$AQI, decreasing = TRUE), ]

print("\n도시별 평균 AQI (상위 10개):")
head(city_avg_aqi, 10)

# AQI 등급별 도시 수
aqi_distribution <- table(air_quality$AQI_Category)
print("\nAQI 등급별 분포:")
print(aqi_distribution)
```
