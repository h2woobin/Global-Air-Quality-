```r
file_path <- "../data/global_air_quality_dataset.csv"

if(file.exists(file_path)) {
  air_quality <- read.csv(file_path)
  head(air_quality)
} else {
  print("Can't fined the file. Please check the path.")
}
```

```r
summary(air_quality)
```

```r
str(air_quality)
```

```r
#Data cleaning
colnames(air_quality) <- c(
  "Date",
  "City",
  "Country",
  "AQI",
  "PM2.5",
  "PM10",
  "NO2",
  "SO2",
  "CO",
  "O3",
  "Temperature",
  "Humidity",
  "WindSpeed"
)

print("Missing values per column:")
print(colSums(is.na(air_quality)))

air_quality$Date <- as.Date(air_quality$Date)
```

```r
air_quality_clean <- air_quality 
head(air_quality_clean)
```

```r
plot(x = air_quality_clean$SO2, y = air_quality_clean$PM2, xlab = "SO2 concentration (ppb)", ylab = "PM2.5 concentration (µg/m³)")
plot(x = air_quality_clean$O3, y = air_quality_clean$PM2, xlab = "O3 concentration (ppb)", ylab = "PM2.5 concentration (µg/m³)")
plot(x = air_quality_clean$CO, y = air_quality_clean$PM2, xlab = "CO concentration (ppm)", ylab = "PM2.5 concentration (µg/m³)")
```

```r
model <- lm(air_quality_clean$PM2 ~ air_quality_clean$SO2 + air_quality_clean$CO + air_quality_clean$O3, data = air_quality_clean)

plot(model$fitted.values, model$residuals, xlab="Fitted value", ylab="Residual")
```

```r
log_PM_model <- lm(log(air_quality_clean$PM2) ~ air_quality_clean$SO2 + air_quality_clean$CO + air_quality_clean$O3, data = air_quality_clean)
summary(log_PM_model)
log_cov_model <- lm(air_quality_clean$PM2 ~ log(air_quality_clean$SO2) + log(air_quality_clean$CO) + log(air_quality_clean$O3), data = air_quality_clean)
summary(log_cov_model)
plot(log_PM_model$fitted.values, log_PM_model$residuals, xlab="Fitted value", ylab="Residual")
plot(log_cov_model$fitted.values, log_cov_model$residuals, xlab="Fitted value", ylab="Residual")

```
